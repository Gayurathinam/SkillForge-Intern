```python
import pandas as pd
df = pd.read_csv('Iris.csv')
x = df["SepalLengthCm"].mean()
df["SepalLengthCm"].fillna(x, inplace = True)
print("Mean:", x)
```

```python
import pandas as pd
df = pd.read_csv('Iris.csv')
x = df["PetalLengthCm"].median()
df["PetalLengthCm"].fillna(x, inplace = True)
print("Median:", x)
```

```python
import pandas as pd
df = pd.read_csv('Iris.csv')
x = df["PetalWidthCm"].mode()[0]
df["PetalWidthCm"].fillna(x, inplace = True)
print("Mode:", x)
```

```python
sns.set(style="whitegrid")

# Histogram
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='SepalLengthCm', kde=True, color='green')
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.set(style="whitegrid")
# Box Plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']])
plt.title('Box Plot of Iris Features')
plt.ylabel('Measurement (cm)')
plt.show()
```

```python
sns.set(style="whitegrid")
# Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='SepalLengthCm', y='PetalLengthCm', hue='Species')
plt.title('Scatter Plot of Sepal Length vs Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()
```

```python
sns.set(style="whitegrid")
# Heatmap (Correlation Matrix)
plt.figure(figsize=(10, 6))
correlation_matrix = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
